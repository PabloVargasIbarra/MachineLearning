```python
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
superficie = np.array([200, 68, 55, 100, 180])
precio = np.array([2375, 1200, 950, 1800, 2000])
one = np.ones([len(superficie),1],float)
X = np.transpose([superficie])
y = np.transpose([precio])
```

```python
reg = linear_model.LinearRegression()
reg.fit(X, y)
reg.coef_, reg.intercept_
```

```python
plt.scatter(superficie, precio)
mapp = np.arange(40,210,1)
plt.plot(mapp, 8.36*mapp + 656, color = "red")
plt.xlabel("superficie")
plt.ylabel("precio")
plt.title("f(x) = 8.36x + 656")
plt.show()
```
